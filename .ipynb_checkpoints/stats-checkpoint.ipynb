```python
data_name = "output/stats_006.dat"
headers = [ 'Flux', 'Abund', 'T', 'z', 'n_H', 'Chnls', '$E_{min}$', '$E_{max}$', 'E_{sum}', 'cs', 'ecs' ] 
table = pd.read_csv(data_name, sep = ' ', names = headers)#[table['Abund'] == 0.2]#[404:501] #0.2
table
```

```python
data_name_mekal = "output/stats_007_mekal.dat"
table1 = pd.read_csv(data_name_mekal, sep = ' ', names = headers)#[table['Abund'] == 0.2]#[404:501] #0.2
table1
```

```python
abundance = table['Abund'].to_numpy().astype(float)[0]
absorption = table['n_H'].to_numpy().astype(float)[0]
redshift = table['z'].to_numpy().astype(float)[0]
```

ABUNDANCE

```python
plt.figure(figsize=(10,7))

for abund in sorted(list(set(table['Abund']))):
    
    data = table[table['Abund'] == abund].to_numpy().astype(float)
    ecs = data[:,10]
    cs = data[:,9]
    e_mean = ecs/cs
    temp = data[:,2]
    
    

    plt.plot(temp, e_mean, label = "Z = "+str(round(abund, 1)))
    
    #plt.plot(temp, data1[:,10]/data1[:,9]/e_mean, linewidth=3, label = "Z = "+str(round(ab, 3)))
    
plt.xlabel('Temperature (keV)', fontsize = 15)
plt.ylabel('Average energy (keV)', fontsize = 15)
plt.title('<E>$_{mekal}$ / <E>$_{apec}$ (T) for $n_H ='+str(absorption)+'\cdot 10^{22} \ cm^{-2}$; z = 0; Z is in Solar units', fontsize = 15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xscale('log')
#plt.yscale('log')
#plt.xlim(0.05, 10)
#plt.ylim(6e-5, 0.8)
plt.legend(fontsize=15, loc = 4)
plt.grid()
#savefig()
plt.show()
```

ABSORPTION

```python
plt.figure(figsize=(14,10))

for ab in sorted(list(set(table['n_H']))):
    
    data = table[table['n_H'] == ab].to_numpy().astype(float)
    ecs = data[:,10]
    cs = data[:,9]
    e_mean = ecs/cs
    temp = data[:,2]

    plt.plot(temp, e_mean, linewidth=2, label = "$n_H$ = "+str(round(ab, 3))+' (Chandra)')
    
    
    #data1 = table1[table1['n_H'] == ab].to_numpy().astype(float)
    #plt.plot(data1[:,2], data1[:,10]/data1[:,9], 
    #         linestyle='-.',
    #         linewidth=2,
    #         label = "$n_H$ = "+str(round(ab, 3))+' (mekal)')
    
    #plt.plot(temp, data1[:,10]/data1[:,9]/e_mean, linewidth=3, label = "$n_H$ = "+str(round(ab, 3)))
    
plt.xlabel('Temperature (keV)', fontsize = 15)
#plt.ylabel('Average energy (keV)', fontsize = 15)
plt.title('<E>(T) for Z = '+str(abundance)+' Solar; z = '+str(redshift)+'; $n_H$ in $10^{22} \ cm^{-2}$', fontsize = 15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(0.08, 10)
#plt.ylim(6e-5, 0.8)
plt.legend(fontsize=15, loc = 0)
plt.grid()
savefig()
plt.show()
```

REDSHIFT: MEKAL vs APEC

```python
plt.figure(figsize=(12,10))

for redshift in sorted(list(set(table['z']))):
    
    data = table[table['z'] == redshift].to_numpy().astype(float)
    ecs = data[:,10]
    cs = data[:,9]
    e_mean = ecs/cs
    temp = data[:,2]
    a = plt.plot(temp, e_mean,
             linestyle='-',
             linewidth=3,
             label = "z = "+str(round(redshift, 2)))#+' (APEC)')
    color1 = a[-1].get_color()
    
    data1 = table1[table1['z'] == redshift].to_numpy().astype(float)
    ecs1 = data1[:,10]
    cs1 = data1[:,9]
    e_mean1 = ecs1/cs1
    temp1 = data1[:,2]
    plt.plot(temp1, e_mean1,
             linestyle=':',
             linewidth=3,
             color = color1)
            #label = "z = "+str(round(redshift, 3))+' (MEKAL)')
    
plt.xlabel('T, keV', fontsize = 15)
plt.ylabel('Average energy, keV', fontsize = 15)
plt.title('<E> (T) for Z ='+str(abundance)+' Solar; $n_H = '+str(absorption)+' \cdot 10^{22} \ cm^{-2}$', fontsize = 15)

plt.xticks(size=15)
plt.yticks(size=15)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(0.08, 11)
#plt.ylim(-1.0, 1.1)

from matplotlib.lines import Line2D

handles, labels = plt.gca().get_legend_handles_labels()
line_a = Line2D([0], [0], label='APEC', color='black', linestyle='-')
line_m = Line2D([0], [0], label='MEKAL', color='black', linestyle=':')
handles.extend([line_a, line_m])

plt.legend(handles=handles, fontsize=16, loc = 4)
plt.grid()
savefig()
plt.show()
```

REDSHIFT: MEKAL / APEC

```python
plt.figure(figsize=(12,10))

for redshift in sorted(list(set(table['z']))):
    
    data = table[table['z'] == redshift].to_numpy().astype(float)
    ecs = data[:,10]
    cs = data[:,9]
    e_mean = ecs/cs
    temp = data[:,2]
    
    data1 = table1[table1['z'] == redshift].to_numpy().astype(float)
    ecs1 = data1[:,10]
    cs1 = data1[:,9]
    e_mean1 = ecs1/cs1
    temp1 = data1[:,2]
    
    plt.plot(temp, e_mean1/e_mean,
             linestyle='-',
             linewidth=3,
             label = "z = "+str(round(redshift, 3)))
    
plt.xlabel('T, keV', fontsize = 15)
plt.ylabel('Average energy <E>, keV', fontsize = 15)
plt.title('<E>$_{mekal}$ / <E>$_{apec}$ (T) for Z ='+str(abundance)+' Solar; $n_H = '+str(absorption)+' \cdot 10^{22} \ cm^{-2}$', fontsize = 15)

plt.xticks(size=15)
plt.yticks(size=15)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(0.08, 11)
#plt.ylim(-1.0, 1.1)
plt.legend(fontsize=16, loc = 0) #title='$E_0 = <E(z=0,T)>$'
plt.grid()
savefig()
plt.show()
```

REDSHIFT SCALING

```python
#plt.figure(figsize=(10,7))
plt.figure(figsize=(14,10))

alpha = 0.0
beta = 0.85

for redshift in sorted(list(set(table['z']))):#[0::2]:
    
    data = table[table['z'] == redshift].to_numpy().astype(float)
    ecs = data[:,10]
    cs = data[:,9]
    e_mean = ecs/cs
    temp = data[:,2]
    a = plt.plot(temp, e_mean,
             linestyle='-',
             linewidth=3,
             label = "z = "+str(round(redshift, 2)))#+' (eROSITA)')
    color1 = a[-1].get_color()
    
    #data1 = table1[table1['z'] == redshift].to_numpy().astype(float)
    #ecs1 = data1[:,10]
    #cs1 = data1[:,9]
    #e_mean1 = ecs1/cs1
    #temp1 = data1[:,2]
    #plt.plot(temp1, e_mean1, linewidth=3, linestyle='-.', label = "z = "+str(round(redshift, 3))+' (Chandra)')
    
    #plt.plot(temp, e_mean1/e_mean, linewidth=3, label = "z = "+str(round(redshift, 3)))
    
    data_z0 = table[table['z'] == 0.0].to_numpy().astype(float)
    plt.plot(data_z0[:,2]*(1+redshift)**alpha, data_z0[:,10]/data_z0[:,9]/(1+redshift)**beta, 
             linestyle = '--',
             linewidth=3,
             color = color1,
             label = '$E_0 / (1+'+str(round(redshift,2))+')^{'+str(beta)+'}$')
    
plt.xlabel('$T \cdot \ (1+z)^{'+str(alpha)+'}$ (keV)', fontsize = 15)
plt.ylabel('Average energy (keV)', fontsize = 15)
#<E>$_{mekal}$ / <E>$_{apec}$
plt.title('<E> (T) for Z ='+str(abundance)+' Solar; $n_H = '+str(absorption)+' \cdot 10^{22} \ cm^{-2}$; $\\alpha='+str(alpha)+', \\beta = '+str(beta)+' $', fontsize = 15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(0.08, 100)
#plt.ylim(-1.0, 1.1)
plt.legend(fontsize=16, loc = 5) #title='$E_0 = <E(z=0,T)>$'
plt.grid()
savefig()
plt.show()
```
