```python
model = pd.read_csv("spectra_xspec/cflux_wabs_apec_1_keV.dat", sep = ' ', skiprows=[0,1,2], header=None)
model = model.to_numpy().astype(float)

plt.figure(figsize=(17,7))

x = model[:,0]
y1 = model[:,1]
y2 = model[:,2]

plt.errorbar(x,y2, linewidth=1)#, xerr=xe, yerr=ye)
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.09, 11.)
#plt.ylim(10**(-6), 0.05)
plt.ylim(10**(-9), 2.0)
plt.xlabel('E, keV', fontsize = 16)
plt.ylabel('F, $phot \ s^{-1} \ cm^{-2} \ keV^{-1}$', fontsize = 16)
plt.title('Model spectrum ($cflux*wabs*apec$)', fontsize = 16)
plt.xticks(size=15)
plt.yticks(size=15)
t1 = [0.2, 0.4, 0.6, 1, 2, 4, 10]
plt.xticks(t1,t1)
plt.grid()
plt.savefig("model_spectrum_cflux_wabs_apec.png")
plt.show()
```

```python
model_cxb = pd.read_csv("spectra_xspec/cxb_model.dat", sep = ' ', skiprows=[0,1,2], header=None)
model = model_cxb.to_numpy().astype(float)

plt.figure(figsize=(15,7))

x = model[:,0]
y1 = model[:,1]
y2 = model[:,2]
y3 = model[:,3]
y4 = model[:,4]
y5 = model[:,5]
y6 = model[:,6]

from scipy.interpolate import spline


xnew = np.linspace(0.1, 2., 100)  
s2 = spline(x, y2, xnew)
s3 = spline(x, y3, xnew)
s4 = spline(x, y4, xnew)
s5 = spline(x, y5, xnew)
s6 = spline(x, y6, xnew)

#plt.plot(x, y1)
plt.plot(x, s2, label='All')
plt.plot(x, s3, label='Local Bubble')
plt.plot(x, s4, label='Galactic halo')
plt.plot(x, s5, label='Hot part')
plt.plot(x, s6, label='AGNs')
#plt.errorbar(x,y2, linewidth=3)#, xerr=xe, yerr=ye)
plt.xscale('log')
plt.yscale('log')
plt.xlim(0.09, 4.)
plt.ylim(5*10**(-7), 0.001)
plt.xlabel('E, keV', fontsize = 16)
plt.ylabel('F, $phot \ s^{-1} \ cm^{-2} \ keV^{-1}$', fontsize = 16)
plt.title('Model: $constant(apec + wabs*(apec + apec + 1*pow))$', fontsize = 16)
plt.xticks(size=15)
plt.yticks(size=15)
plt.xticks(t1,t1)
plt.grid()
plt.legend(fontsize=16)
#plt.savefig("model_cxb.png")
plt.show()
```
